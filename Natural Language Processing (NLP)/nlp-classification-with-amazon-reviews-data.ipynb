# Feature Engineering

```python
del df["reviewerID"]
del df["asin"]
del df["reviewerName"]
del df["unixReviewTime"]
```

```python
df.isnull().sum()
```

```python
df['text'] = df['summary'] + ' ' + df['reviewText']
```

```python
del df["summary"]
del df["reviewText"]
```

```python
df["text"]=df["text"].str.lower()
df["text"]=df["text"].str.replace("[^\w\s]","")
#noktalama işaretlerini kaldırır
df["text"]=df["text"].str.replace("\d+","")
#rakamları kaldırır
df["text"]=df["text"].str.replace("\r","").replace("\n","")
#satır başları, boşlukları kaldır.
```

```python
import html

text = df["text"]
decoded_text = html.unescape(text)
```

```python
df["text"]=decoded_text
```

```python
df["text"]
```

```python
import nltk
from nltk.corpus import stopwords

# Stopwords listesini indirin
nltk.download('stopwords')

# Stopwords listesini alın
stopwords_list = stopwords.words('english')

# Cümlelerin olduğu bir sütununuz olsun
df['text'] = df['text'].apply(lambda x: ' '.join([word for word in x.split() if word.lower() not in stopwords_list]))
```
