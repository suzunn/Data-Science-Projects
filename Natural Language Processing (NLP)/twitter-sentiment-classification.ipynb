```python
import pandas as pd
```

```python
df=pd.read_csv("/kaggle/input/twitter/train.csv")
```

# EDA

```python
df.head()
```

```python
df.isnull().sum()
```

# Feature Engineering

```python
df=df[df['text'].notnull()]
```

```python
df=df[["text","sentiment"]]
```

```python
df["text"]=df["text"].str.lower()
df["text"]=df["text"].str.replace("[^\w\s]","")
#noktalama işaretlerini kaldırır
df["text"]=df["text"].str.replace("\d+","")
#rakamları kaldırır
df["text"]=df["text"].str.replace("\r","").replace("\n","")
#satır başları, boşlukları kaldır.
```

```python
df["sentiment"].unique()
```

```python
pozveneg=df[ (df['sentiment']=="positive") | (df['sentiment']=="negative") ]
#iyi ve kotu olanları, 1 yıldız alan ve 5 yıldız alanlara eşitledik
```

```python
pozveneg.reset_index(drop=True,inplace=True)
```

```python
x=pozveneg["text"]
y=pozveneg["sentiment"]
```

# Modelling

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
vect=CountVectorizer(ngram_range=(1,2),stop_words="english")
                     #ngram= kelimeleri 1,1; 2,2 okuyarak bidaha dene
```

```python
x=vect.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=20,random_state=42)
```

```python
tf=pd.DataFrame(x_train.toarray(),columns=vect.get_feature_names_out())
```

```python
nb=MultinomialNB()
```

```python
model=nb.fit(x_train,y_train)
```

```python
pred=nb.predict(x_test)
```

```python
accuracy_score(pred,y_test)
```
