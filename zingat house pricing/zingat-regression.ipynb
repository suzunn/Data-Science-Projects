# Feature Engineering

```python
del df["Unnamed: 0"]
```

```python
df["BRÜT(m²)"].unique()
```

```python
df["BRÜT(m²)"] = pd.to_numeric(df["BRÜT(m²)"], errors="coerce")
df["BRÜT(m²)"].fillna(0, inplace=True)
```

```python
df["NET(m²)"].unique()
```

```python
df["NET(m²)"] = df["NET(m²)"].replace("75m²", "75")
df["NET(m²)"] = df["NET(m²)"].replace("720m²", "720")
df["NET(m²)"] = pd.to_numeric(df["NET(m²)"], errors="coerce")
df["NET(m²)"].fillna(0, inplace=True)
```

```python
df["NET(m²)"] = df["NET(m²)"].replace("m²", "")
df["BRÜT(m²)"] = pd.to_numeric(df["BRÜT(m²)"], errors="coerce")
#errors="coerce" sayısal olmayan değerleri NaN ile değitirir
df["BRÜT(m²)"].fillna(0, inplace=True)
```

```python
df["FOTOÐRAF SAYISI"].fillna(0, inplace=True)
```

```python
df["ODA-SALON SAYISI"].unique()
```

```python
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("2+1", "3")
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("3+1", "4")
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("1+1", "2")
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("345", "5")
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("4+1", "5")
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].replace("0 TL", "0")
```

```python
df["ODA SAYISI"].unique()
```

```python
df["ODA SAYISI"] = df["ODA SAYISI"].replace("0 TL", "0")
df["ODA SAYISI"] = df["ODA SAYISI"].replace("345", "5")
```

```python
df["BANYO SAYISI"].unique()
```

```python
df["BANYO SAYISI"] = df["BANYO SAYISI"].replace("345 (m²)", "345")
df["BANYO SAYISI"] = df["BANYO SAYISI"].replace("21-25 arasý", "23")
df["BANYO SAYISI"] = df["BANYO SAYISI"].replace("Hayýr", "0")
```

```python
df["FOTOÐRAF SAYISI"].unique()
```

```python
df["FÝYAT"] = df["FÝYAT"].astype(int)
df["NET(m²)"] = df["NET(m²)"].astype(int)
df["BRÜT(m²)"] = df["BRÜT(m²)"].astype(int)
df["ODA-SALON SAYISI"] = df["ODA-SALON SAYISI"].astype(int)
df["ODA SAYISI"] = df["ODA SAYISI"].astype(int)
df["BANYO SAYISI"] = df["BANYO SAYISI"].astype(int)
df["FOTOÐRAF SAYISI"] = df["FOTOÐRAF SAYISI"].astype(int)
```

```python
df["FÝYAT"]=df["FÝYAT"]*1000
```

```python
df.info()
```

```python
df.head()
```

```python
new_df = pd.get_dummies(df, columns=["MAHALLE"])
new_df.drop(columns="ODA-SALON SAYISI", inplace=True)
new_df.drop(columns="ÝLAN ADI", inplace=True)
new_df = new_df.astype(int)
new_df
```

```python
new_df.sort_values("FÝYAT")
```

# Modelling

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import RidgeCV
import statsmodels.api as sm
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
from pandas import Series, DataFrame
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
import patsy
```

```python
x = new_df.iloc[:,1:]
y = new_df.iloc[:,:1]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=20)
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.25, random_state=20)

lr_model = LinearRegression()

lr_model.fit(x_train,y_train)
pred = lr_model.predict(x_val)

print("MSE Value: ", np.mean((pred - y_val)**2), "\n")

print("R^2 Score Value (Test): ", lr_model.score(x_test, y_test), "\n")

print("R^2 Score Value (Validation): ", lr_model.score(x_val, y_val), "\n")

print("Coefficients:\n", lr_model.coef_)
```
