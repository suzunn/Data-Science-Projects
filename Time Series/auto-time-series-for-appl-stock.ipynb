```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("AAPL.csv")
```

```python
df.head()
```

```python
df=df[["Date","Close"]]

df["Date"]= pd.to_datetime(df.Date)
#date'i datetime yapma

df["Close"].plot(figsize=(12, 8), title= "Apple Stok Fiyatları", fontsize=20, label="Kapanış Fiyatı")

plt.legend()
#grafiğin açıklama etiketinin görüntülenmesini sağlar

plt.grid()
#grafiğin üzerinde ızgaranın görüntülenmesini sağlar

plt.show()
```

```python
from autots import AutoTS

model = AutoTS(forecast_length=10, frequency='infer',   ensemble='simple',    drop_data_older_than_periods=200)
               #10 adet tahmin     #frekansı çıkartır   #tahmin birleştirme   #200 periyottan eskiler atilsin

model = model.fit(df, date_col='Date', value_col='Close', id_col=None)
```

```python
prediction = model.predict()
forecast = prediction.forecast
print("Apple'ın Stok Fiyat Tahmini")
print(forecast)
```
